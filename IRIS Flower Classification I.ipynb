```python
# Inspect the Data.

setosa = iris[iris['Species'] == 'Iris-setosa']
versicolor = iris[iris['Species'] == 'Iris-versicolor']
virginica = iris[iris['Species'] == 'Iris-virginica']

plt.scatter(setosa['SepalLengthCm'], setosa['SepalWidthCm'], label='Iris-setosa', marker='o', color='blue')
plt.scatter(versicolor['SepalLengthCm'], versicolor['SepalWidthCm'], label='Iris-versicolor', marker='^', color='green')
plt.scatter(virginica['SepalLengthCm'], virginica['SepalWidthCm'], label='Iris-virginica', marker='s', color='red')

plt.xlabel("SepalLength(CM)")
plt.ylabel("Sepal Width(CM)")
plt.title("SepalLength vs SepalWidth")
plt.legend()
plt.show()
```

```python
# Split the Data into Training & Testing Sets.

from sklearn.model_selection import train_test_split
IRIS = iris.drop(['Id',"Species"],axis=1)
iris_feature = np.array(IRIS)
iris_label = np.array(iris['Species'])
x_train,x_test,y_train,y_test = train_test_split(iris_feature, iris_label, random_state=10,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
# Train the Model using KNN

from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=1)
KNN.fit(x_train,y_train)
```

```python
# Make Predictions

import numpy as np

new_data = np.array([[6.4,2.7,5.3,1.9]])
new_data_DF = pd.DataFrame(new_data, columns=IRIS.columns)
new_data_DF
```

```python
y_pred = KNN.predict(x_test)
pred_DF = pd.DataFrame(x_test,columns=IRIS.columns)
pred_DF['Predicted Species'] = y_pred
pred_DF
```
